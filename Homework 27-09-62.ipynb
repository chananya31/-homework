```python
"""
โจทย์การใช้ if
ไม้ท่อนหนึ่งยาว 8.25 เมตร และอีกท่อนหนึ่งยาว 4.3 เมตร มาเชื่อมกันแล้วตัดออกไปทำขาโต๊ะ 3 เมตร จะเหลืออีกกี่เมตร
"""
Timber1 = float(input("ไม้ท่อนที่ 1 ยาว : "))
Timber2 = float(input("ไม้ท่อนที่ 2 ยาว : "))
cut     = float(input("ตัดไปทำขาโต๊ะ 1 ขา ยาว : "))
Mix     = Timber1 + Timber2
Timber   = Mix - cut
print("จะเหลือไม้อีก = ",Timber,"เมตร")
if Timber > 10 :
    print(" Use Timber  ")
```

```python
"""
โจทย์การใช้ if else elif
แม่ค้าขายผลไม้ไป 20 กิโลกรัม ราคากิโลกรัมละ 15 บาท เอาเงินที่ขายได้ไปซื้อแครอท 
ราคาหัวละ 10 บาท แม่ค้าจะซื้อแครอทได้ประมาณกี่หัว *กำหนดให้ได้แอปเปิ้ล 30 หัว *
"""
V = int(input("ขายผลไม้ไป : "))
S = int(input("ราคากิโลกรัมละ : "))
P = int(input("แครอทหัวละ : "))
money = V * S
count = int(money // P)
print(" จะได้แครอทประมาณ ",count, "หัว")
if count < 1 :
    print("ไม่สามารถซื้อได้")
elif count > 35 :
    print("เกินจำนวนที่ต้องการ")
else:
    count == 30
    print("ได้ครบตามจำนวน")
```

```python
"""
โจทย์การใช้ while loop
โปรโมชั่นทางร้าน ซื้ออาหารมากกว่า 2,000 บาท ขึ้นไป ลดราคา 10 เปอร์เซ็น
------ แม่ค้าร้านอาหาร ------
คิดเงินค่าอาหารให้ลูกค้าโต๊ะที่ 1 รวมเงิน 300 บาท ต้องชำระเงินกี่บาท
คิดเงินค่าอาหารให้ลูกค้าโต๊ะที่ 2 รวมเงิน 3,500 บาท ต้องชำระเงินกี่บาท

"""
while(True):
    Number = int(input("ยอดเงินรวม = "))
    if(Number<2000):
        print("ต้องชำระเงินจำนวน = ",Number,"บาท")
    else:
        Number > 2000
        sale = round(Number * 0.10)
        pay  = Number - sale
        print("คิดราคาโปรโมชั่นต้องชำระเงินจำนวน = ",pay,"บาท")
        break
```

```python
"""
โจทย์การใช้ for loop
ร้านค้าแห่งหนึ่ง ขายข้าวสาร ราคากิโลกรัม 30 บาท แต่ซื้อได้ไม่เกินครอบครัวล่ะ 10กิโลกรัม
โดยติดป้ายราคา ตั้งแต่ถุงที่1 ถึงถุงที่10 ไว้ดังนี้ ให้คำนวณราคาตั้งแต่ถุงที่1-10 ว่ามีราคาเท่าไหร่บ้าง
"""
Price = int(input("ราคาข้าวสารกิโลกรัมละ = "))
print("ราคา x กิโลกรัม ")
for i in range(1,10):
    print(Price,"*" , i,"=",Price*i, "บาท")
```

```python
"""
ในการเทปูนพื้นห้องกว้าง 4.5 เมตร ยาว 8.75 เมตร เสียค่าใช้จ่ายตารางเมตรละ 320 บาท
อยากทราบว่าจะต้องเสียเงินในการเทคอนกรีตทั้งหมดกี่บาท
"""
wide  = float(input("พื้นห้องกว้าง = "))
long  = float(input("พื้นห้องยาว  = "))
po   = int(input("จ่ายตารางเมตรละ = "))
mix   = wide * long
print("พื้นที่ห้องมี",mix, "ตารางเมตร")
pay = mix * po 
print("ต้องเสียเงินในการเทปูนกรีต",pay,"บาท")
```

```python
"""
ไม้หนึ่งยาว 15 เมตร ส่วนที่อยู่ในโคลน 4 เมตร
และส่วนที่อยู่ในน้ำยาว 3 เมตร ส่วนที่พ้นน้ำจะยาวกี่เมตร
"""
L = int(input("ไม้ท่อนหนึ่งยาว : "))
A = int(input("ส่วนที่อยู่ในโคลน : "))
O = int(input("ส่วนที่อยู่ในน้ำ   : "))
print("ส่วนที่พ้นน้ำจะต้องยาว ",(L-(A+O)),"เมตร")
```

```python
"""
นักเรียนห้องหนึ่งมีจำนวน 40 คน ไม่มาเรียนจำนวน  0.5 ของนักศึกษาทั้งหมด จงหานักศึกษาที่ไม่มาเรียน
"""
count = int(input("จำนวนนักเรียนทั้งหมด : "))
absent = input("ไม่มาเรียน : ")
m       = round(int(count) *float(absent))
print("นักศึกษาที่ไม่มาเรียน",m,"คน")
```

```python
"""
วาสนาซื้อแหวนเพชรคอราคา 25,000.0 บาท ขายไปและขาดทุน 0.20 ของราคาสร้อย  อยากทราบว่าวาสนาขายแหวนเพชรไปราคากี่บาท
"""
Price = float(input("แหวนเพชรราคา : "))
Sale  = float(input("ขาดทุน : "))
print("วาสนาขายสร้อยไปในราคา",int(Price*Sale),"บาท")
```

```python
"""
สมมติว่าเดือนหน้า เราจะเดินทางไปนิวซีเเลน มีเงินไทยอยู่ 9000.0 บาท จะแลกได้เท่าไร
*กำหนดให้ใช้สกุลเงิน NZD โดย 1 NZD = 19.6770 บาทไทย
"""
Thi = float(input("จำนวนเงินไทยที่ต้องการแลก : "))
NZD = float(input(" ค่า 1 NZD :  "))
print("จะแลกได้",((Thi*1)//NZD),"NZD")
```
